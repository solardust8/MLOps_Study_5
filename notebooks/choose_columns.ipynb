# Read dataset

```python
df = pd.read_csv("../data/small_openfoods.csv", sep='\t')
```

# Find columns without NaN values

```python
def find_columns_with_no_na(df: pd.DataFrame):
    return set(df.columns[df.notna().all()].to_list())
```

```python
cols = find_columns_with_no_na(df)
```
